```python
# This code is for splitting the pdf pages into jpg images 
import fitz
from PIL import Image

def convert_pdf_to_images(pdf_path, output_folder):
    # Open the PDF file
    pdf_document = fitz.open(pdf_path)

    # Iterate through each page in the PDF
    for page_num in range(pdf_document.page_count):
        # Get the current page
        pdf_page = pdf_document[page_num]

        # Convert the PDF page to an image
        image = pdf_page.get_pixmap()

        # Create a PIL image from the raw image data
        pil_image = Image.frombytes("RGB", (image.width, image.height), image.samples)

        # Save the image
        image_path = f"{output_folder}/page_{page_num + 1}.jpg"
        pil_image.save(image_path, 'JPEG')

        print(f"Saved: {image_path}")

    # Close the PDF document
    pdf_document.close()

# Specify the path to the input PDF file
input_pdf_path = "C:/Users/ANURAG JADHAV/OneDrive/Documents/DS/project file/Images/7.pdf"

# Specify the folder where the output images will be saved
output_folder = "C:/Users/ANURAG JADHAV/OneDrive/Documents/DS/project file/pdf split/7"

# Create the output folder if it doesn't exist
import os
os.makedirs(output_folder, exist_ok=True)

# Convert the PDF to images
convert_pdf_to_images(input_pdf_path, output_folder)
```

```python
# This code crops the pdf split images to a broder less image 
import cv2
import os

def save_cropped_signature(image_path, output_folder):
    # Load the image
    image = cv2.imread(image_path, cv2.IMREAD_COLOR)

    # Convert to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply a Gaussian blur
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)

    # Use thresholding to create a binary image
    _, binary = cv2.threshold(blurred, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

    # Find contours
    contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Filter contours based on size and aspect ratio
    signature_contours = [cnt for cnt in contours if cv2.contourArea(cnt) > 100]  # Example threshold value

    # Assuming we want to save the first signature detected
    if signature_contours:
        x, y, w, h = cv2.boundingRect(signature_contours[0])
        # Crop the signature
        cropped_signature = image[y:y+h, x:x+w]

        # Save the cropped signature image to the specified output folder
        os.makedirs(output_folder, exist_ok=True)
        output_path = os.path.join(output_folder, 'det_signature.jpg')
        cv2.imwrite(output_path, cropped_signature)
        print(f"Cropped signature saved as '{output_path}'")
    else:
        print("No signature detected.")

# Specify the path to the image
image_path = "C:/Users/ANURAG JADHAV/OneDrive/Documents/DS/project file/pdf split/7/page_5.jpg"  # Replace with the actual path to your image

# Specify the output folder
output_folder = "C:/Users/ANURAG JADHAV/OneDrive/Documents/DS/project file/cropped border/7"

# Call the function to save the cropped signature image
save_cropped_signature(image_path, output_folder)
```

```python
#This code is for extracting the signatures
import cv2
import os

# Load the image
image_path = "C:/Users/ANURAG JADHAV/OneDrive/Documents/DS/project file/cropped border/7/page_3.jpg"  # Make sure this path is correct
image = cv2.imread(image_path)

# Check if the image is loaded properly
if image is None:
    print(f"Error: Unable to load image at {image_path}")
    exit()

# Define the number of rows and columns in the grid
num_rows = 4
num_columns = 2

# Calculate the approximate width and height of each cell containing a signature
cell_width = image.shape[1] // num_columns
cell_height = image.shape[0] // num_rows

# Specify the output folder
output_folder = "C:/Users/ANURAG JADHAV/OneDrive/Documents/DS/project file/cropped signatures/7/page 3"
os.makedirs(output_folder, exist_ok=True)

# Function to crop the signature based on the cell position
def crop_signature(row, col, cell_width, cell_height, source_image, margin=10):
    # Calculate start and end points, subtracting the margin
    start_x = col * cell_width + margin
    start_y = row * cell_height + margin
    end_x = (col + 1) * cell_width - margin
    end_y = (row + 1) * cell_height - margin
    # Make sure the end points are within the image boundaries
    end_x = min(end_x, source_image.shape[1])
    end_y = min(end_y, source_image.shape[0])
    return source_image[start_y:end_y, start_x:end_x]

# Crop each cell and save the signature as JPG
for row in range(num_rows):
    for col in range(num_columns):
        cropped_signature = crop_signature(row, col, cell_width, cell_height, image)
        signature_file_path = os.path.join(output_folder, f'signature_row{row}_col{col}.jpg')
        cv2.imwrite(signature_file_path, cropped_signature)
        print(f"Saved: {signature_file_path}")

print("All signatures have been cropped and saved as JPG files.")
```

```python
# This code augments the signatures 
from PIL import Image, ImageEnhance, ImageOps, ImageFilter
import os
import numpy as np
def augment_images(input_folder, output_folder):
    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    # List all files in the input folder
    files = os.listdir(input_folder)
    for file in files:
        # Check if the file is an image
        if file.endswith(".jpg") or file.endswith(".jpeg") or file.endswith(".png"):
            # Open the image file
            image_path = os.path.join(input_folder, file)
            img = Image.open(image_path)
            # Augmentation list
            augmented_images = []
            # Rotations at different angles
            for angle in [45, 90, 135]:
                rotated = img.rotate(angle, expand=1, fillcolor='white')
                augmented_images.append(rotated)
            # Brightness and contrast enhancement
            for enhancement_factor in [0.5, 1.5, 2.0]:
                brightness_enhanced = ImageEnhance.Brightness(img).enhance(enhancement_factor)
                contrast_enhanced = ImageEnhance.Contrast(brightness_enhanced).enhance(enhancement_factor)
                augmented_images.append(contrast_enhanced)
            # Horizontal and vertical flip
            augmented_images.append(img.transpose(Image.FLIP_LEFT_RIGHT))
            augmented_images.append(img.transpose(Image.FLIP_TOP_BOTTOM))
            # Shearing
            sheared = img.transform(img.size, Image.AFFINE, (1, 0.5, 0, 0.5, 1, 0), fillcolor='white')
            augmented_images.append(sheared)
            # Gaussian noise
            np_img = np.array(img)
            for _ in range(2):  # Adding noise twice with different intensities
                noise = np.random.normal(0, 25, np_img.shape)
                noisy_img = Image.fromarray(np.clip(np_img + noise, 0, 255).astype(np.uint8))
                augmented_images.append(noisy_img)
            # Random cropping
            for _ in range(2):  # Cropping the image twice with different areas
                crop_area = (10, 10, img.width - 10, img.height - 10)
                cropped_img = img.crop(crop_area)
                augmented_images.append(cropped_img)
            # Save augmented images
            base_filename, file_extension = os.path.splitext(file)
            for i, augmented_img in enumerate(augmented_images[:10]):  # Limit to 10 images
                output_path = os.path.join(output_folder, f"{base_filename}_aug_{i+1}{file_extension}")
                augmented_img.save(output_path, "JPEG")
if __name__ == "__main__":
    # Specify your input and output folders
    input_folder = "/Users/sriharsha/Documents/Data Set/Sign_9"
    output_folder = "/Users/sriharsha/Documents/agumented images 9"
    augment_images(input_folder, output_folder)
```

```python
# This code renames the images in a folder
import os
 
def rename_images(folder_path):
    """
    Rename images in the specified folder in the format "sign1_1", "sign1_2", etc.
    """
    count = 1
 
    for filename in os.listdir(folder_path):
        if filename.lower().endswith(('.jpg', '.jpeg', '.png')):
            extension = os.path.splitext(filename)[1]
            new_name = f"sign2_{count}{extension}"
            current_path = os.path.join(folder_path, filename)
            new_path = os.path.join(folder_path, new_name)
            os.rename(current_path, new_path)
            count += 1
 
if __name__ == "__main__":
    # Specify the path to your folder containing images
    folder_path = "/Users/sriharsha/Documents/Data Set/Sign_2"
 
    # Rename images in the specified folder
    rename_images(folder_path)
```
